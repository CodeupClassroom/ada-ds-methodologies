```python
sales = pd.read_csv('untidy_data/sales.csv')
sales
```

```python
melt = sales.melt(id_vars='Product', var_name='year_and_measure')
melt.head()
```

```python
year_and_measure_df = melt.year_and_measure.str.split(' ', expand=True)
year_and_measure_df.columns = ['year', 'measure']
year_and_measure_df.head()
```

```python
df = pd.concat([melt, year_and_measure_df], axis=1)
df.drop(columns='year_and_measure', inplace=True)
df.head()
```

```python
df.set_index(['Product', 'year', 'measure']).unstack(2)
```

```python
df.head()
```

```python
sales_almost_tidy = df.pivot_table(values='value', index=['Product', 'year'], columns='measure')
```

```python
sales_almost_tidy.head()
```

```python
sales_almost_tidy.columns.name = ''
sales_tidy = sales_almost_tidy.reset_index()
```

```python
sales_tidy
```

```python
cakes = (pd.read_csv('untidy_data/cake_recipes.csv')
 .melt(id_vars=['recipe:position'], var_name='temp', value_name='tastiness'))
cakes.head()
```

```python
cols = cakes['recipe:position'].str.split(':', expand=True)
cols.columns = ['recipe', 'position']
cols.head()
```

```python
cakes = pd.concat([cakes.drop(columns='recipe:position'), cols], axis=1)
```

```python
cakes.groupby('temp').tastiness.mean()
```

```python
cakes.groupby('recipe').tastiness.mean()
```

```python
cakes.sort_values(by='tastiness', ascending=True).tail(1)
```
