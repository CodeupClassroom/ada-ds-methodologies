```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://codeup.com/why-san-antonio-has-more-than-tacos-to-offer/'

headers = {'User-Agent': 'Codeup Ada Data Science'}
response = requests.get(url, headers=headers)
print(response.text)
```

```python
soup = BeautifulSoup(response.text)

# .find - to find a specific element
# .find_all - to find multiple elements matching a criteria

len(soup.find_all(class_='mk-single-content'))

article_content = soup.find(class_='mk-single-content').text
```

```python
with open('article_content.txt', 'w') as f:
    f.write(article_content)
```
